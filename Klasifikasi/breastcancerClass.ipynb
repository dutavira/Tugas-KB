# Classification Breast Cancer Dataset

## import library yang diperlukan

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
```

## Membaca dataset dari file .csv

```python
df = pd.read_csv('breast-cancer.csv')
df
```

## Preprocessing Data

### Encoding
Membuat kelas OrdinalEncoder ke variabel encoder untuk melakukan encoding ordinal kepada kolom yang bertipe objek agar bisa dilakukan learning

```python
encoder = OrdinalEncoder()
```

Melakukan encoding ke kolom bertipe objek

```python
df['Class'] = encoder.fit_transform(df[['Class']])
df['age'] = encoder.fit_transform(df[['age']])
df['menopause'] = encoder.fit_transform(df[['menopause']])
df['tumor-size'] = encoder.fit_transform(df[['tumor-size']])
df['inv-nodes'] = encoder.fit_transform(df[['inv-nodes']])
df['node-caps'] = encoder.fit_transform(df[['node-caps']])
df['breast'] = encoder.fit_transform(df[['breast']])
df['breast-quad'] = encoder.fit_transform(df[['breast-quad']])
df
```

### Data Cleaning

Mengecek kolom yang memiliki nilai null

```python
df.isna().sum()
```

## Supervised Learning

### Menentukan Data Training dan Data Testing

Menentukan kolom mana saja yang dijadikan data learning dan kolom mana yang dijadikan label output

```python
x_data = df[['Class','age','menopause','tumor-size','inv-nodes','node-caps','deg-malig','breast','breast-quad']]
y_label = df[['irradiat']]
```

Melakukan splitting antara data training dan data testing dengan pemisahan sebesar 70:30

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_label, test_size=0.3)
```
